```python
#Save the data with only relevant columns
clean_physiciangeneraldata=renamed_df[['NPI', 'Individual Pac ID','Last Name','First Name','Gender', 'Medical School', 'Graduation Yr', 'Primary Speciality','Organization Name',
       'Organization Pac ID', 'No. Org Memebers', 'Address ln 1','Address ln 2', ' ln_2_sprs', 'City', 'State', 'Zip', 'Phone No.']]
clean_physiciangeneraldata
```

```python
#Load Physician Data file with MIPS score
physicianMISPdata=DataLoader.LoadFile(FilePhysicianIndivialMIPSData)
```

```python
#Display Physician Data file with MIPS score
physicianMISPdata
```

```python
#Merge formatted Physician General Data file and Phisician MIPS Data file
merge_physiciandata_MIPS = pd.merge(clean_physiciangeneraldata,physicianMISPdata, on="NPI", how ="inner")
merge_physiciandata_MIPS.head()
```

```python
#Create dataframe with combined data files for relevant columns
test_physiciandata = merge_physiciandata_MIPS[['NPI', 'Individual Pac ID', 'Last Name_x', 'First Name_x', 'Gender',
       'Medical School', 'Graduation Yr', 'Primary Speciality',
       'Organization Name', 'Organization Pac ID', 'No. Org Memebers',
       'Address ln 1', 'Address ln 2', ' ln_2_sprs', 'City', 'State', 'Zip',
       'Phone No.', 'Quality Category Score', 'ACI Category Score',
       'IA Category Score', 'Final MIPS Score']]
test_physiciandata
```

```python
#Display merged columns
test_physiciandata.columns
```

```python
indexNames = test_physiciandata[test_physiciandata['Final MIPS Score'] == 0 ].index
 
# Delete these row indexes from dataFrame
test_physiciandata.drop(indexNames , inplace=True)
test_physiciandata
```

```python
Rename columns
final_physiciandata=test_physiciandata.rename(columns={"Last Name_x":"Last Name","First Name_x":"First Name"})
final_physiciandata
```
